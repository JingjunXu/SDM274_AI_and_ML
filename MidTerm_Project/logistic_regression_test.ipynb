```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
import mock_data 
import utilities as tool
import logistic_regression as log_reg
import importlib

importlib.reload(mock_data)
importlib.reload(log_reg)
importlib.reload(tool)

# 
data = mock_data.MockData()
dataSet1, labels1 = data.create_linear_data()
dataSet2, labels2 = data.create_quatic_data()
dataSet3, labels3 = data.create_stochastic_data()
train_input1, test_input1, train_target1, test_target1 = train_test_split(dataSet1, labels1, test_size = 0.2)
train_input2, test_input2, train_target2, test_target2 = train_test_split(dataSet2, labels2, test_size = 0.2)
train_input3, test_input3, train_target3, test_target3 = train_test_split(dataSet3, labels3, test_size = 0.2)

# lr
input_dim = 2
batch_size = 1
epoch_num = 1000
tol = None
update_type = 0
iteration = [i for i in range(1, epoch_num + 1)]
model = log_reg.LogisticRegression(input_dim)

# 
learning_rate = 0.01
print('linear_data training')
model.train(train_input1, train_target1, epoch_num, learning_rate, tol, update_type, batch_size)
print('Accuracy for linear_data')
print(accuracy_score(test_target1, model.predict(test_input1)))
print('Precision for linear_data')
print(precision_score(test_target1, model.predict(test_input1)))
print('Recall for linear_data')
print(recall_score(test_target1, model.predict(test_input1)))
print('f1_score for linear_data')
print(f1_score(test_target1, model.predict(test_input1)))
model.plot_loss()
tool.plot_decision_boundary(np.array(dataSet1), np.array(labels1), model)

# 
learning_rate = 0.001
print('quadratic_data training')
model.train(train_input2, train_target2, epoch_num, learning_rate, tol, update_type, batch_size)
print('Accuracy for quatic_data')
print(accuracy_score(test_target2, model.predict(test_input2)))
print('Precision for quatic_data')
print(precision_score(test_target2, model.predict(test_input2)))
print('Recall for quatic_data')
print(recall_score(test_target2, model.predict(test_input2)))
print('f1_score for quatic_data')
print(f1_score(test_target2, model.predict(test_input2)))
model.plot_loss()
tool.plot_decision_boundary(np.array(dataSet2), np.array(labels2), model)

# 
learning_rate = 0.001
print('stochastic_date training')
model.train(train_input3, train_target3, epoch_num, learning_rate, tol, update_type, batch_size)
print('Accuracy for stochastic_date')
print(accuracy_score(test_target3, model.predict(test_input3)))
print('Precision for stochastic_date')
print(precision_score(test_target3, model.predict(test_input3)))
print('Recall for stochastic_date')
print(recall_score(test_target3, model.predict(test_input3)))
print('f1_score for stochastic_date')
print(f1_score(test_target3, model.predict(test_input3)))
model.plot_loss()
tool.plot_decision_boundary(np.array(dataSet3), np.array(labels3), model)
```
