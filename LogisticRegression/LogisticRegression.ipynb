* Use the mult-class logistic regression model to address the multi-class classification for hand-written digits. The training data set is in optdigits.tra and the test data set is optdigits.tes. In both training and test set, the columns from 0 to 63 are input features, representing the gray values of 8-by-8 pixels of an image, while the column 64 refers to the output category (labelled from 0 to 9)

```python
import numpy as np
import multi_class_logistic_regression as MClr
from sklearn.metrics import accuracy_score
import importlib

importlib.reload(MClr)

def getData(file_path):
    input_data = []
    target_data = []
    with open(file_path, 'r') as file:
        for line in file:
            # 通过逗号将每行的数据分别读取，同时将它们都转化成int的数据类型
            int_list = [int(num) for num in line.strip().split(',')]
            # 将输入的特征和标签分隔开来
            input_data.append(int_list[:-1])  
            target_data.append(int_list[-1]) 
    return input_data, target_data

# 读取数据
train_input, train_target = getData('optdigits.tra')
test_input, test_target = getData('optdigits.tes')

# 设置基本参数
class_num = len(np.unique(train_target, return_counts=False))
_, feature_num = np.array(train_input).shape
epoch_num = 500
learning_rate = 0.001

# 构建与训练模型
model = MClr.LogisticRegression(class_num, feature_num)
model.train(train_input, train_target, epoch_num, learning_rate, None, 0, 1)
model.plot_loss()

# 预测测试集的数据以及为其准确性打分
accuracy = accuracy_score(test_target, model.predict(np.array(test_input)))
print(accuracy)
```
